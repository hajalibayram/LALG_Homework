# Improving clustering performance using SVD

```python
# import required functions and libraries
from sklearn.datasets import make_circles
from sklearn.neighbors import kneighbors_graph
from sklearn.cluster import SpectralClustering
import numpy as np
from sklearn.cluster import KMeans
s=289963
np.random.seed(s)
import matplotlib.pyplot as plt
X, labels = make_circles(n_samples=500, noise=0.1, factor=.2)
print('The raw scattered dat')
fig1, ax = plt.subplots()
ax.scatter(X[:, 0], X[:, 1])
fig1.savefig('fig1.eps', format='eps')
```

# KMeans

```python
np.random.seed(s)
kmeans = KMeans(n_clusters=2, random_state=s).fit(X)
print('Clustering results using KMeans:')
fig2, ax = plt.subplots()
ax.scatter(X[:, 0], X[:, 1], c=kmeans.labels_)
fig2.savefig('fig2.eps', format='eps')
```

# Spectral Clustering

```python
np.random.seed(s)
s_cluster = SpectralClustering(n_clusters = 2, eigen_solver='arpack',
        affinity="nearest_neighbors").fit_predict(X)
print('Clustering')
fig3, ax = plt.subplots()
ax.scatter(X[:, 0], X[:, 1], c = s_cluster)
fig3.savefig('fig3.eps', format='eps')
```
